```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import os
import shutil
from fastai.vision.all import *
from scipy.fft import fft, ifft
from scipy import interpolate
from scipy.signal import butter, filtfilt

srate = 220  # Sampling rate

# ----------- Filtering Function ----------- #
def bandpass_filter(signal, lowcut=1.0, highcut=40.0, fs=220, order=5):
    nyq = 0.5 * fs
    b, a = butter(order, [lowcut / nyq, highcut / nyq], btype='band')
    return filtfilt(b, a, signal)

# ----------- Resample Function ----------- #
def resample_array(array, target_length):
    input_indices = np.linspace(0, len(array) - 1, len(array))
    resampled_indices = np.linspace(0, len(array) - 1, target_length)
    interpolator = interpolate.interp1d(input_indices, array)
    return interpolator(resampled_indices).tolist()

# ----------- Create Morlet Wavelets ----------- #
def get_cmwX(nData, freqrange=[1, 6], numfrex=10):
    pi = np.pi
    wavtime = np.arange(-2, 2 - 1/srate, 1/srate)
    nKern = len(wavtime)
    nConv = nData + nKern - 1
    frex = np.linspace(freqrange[0], freqrange[1], numfrex)
    cmwX = np.zeros((numfrex, nConv), dtype=complex)
    numcyc = np.linspace(3, 15, numfrex)

    for fi in range(numfrex):
        s = numcyc[fi] / (2 * pi * frex[fi])
        cmw = np.exp(2j * pi * frex[fi] * wavtime) * np.exp((-wavtime**2) / (2 * s**2))
        cmwX[fi, :] = fft(cmw, nConv) / max(fft(cmw, nConv))
    return cmwX, nKern, frex

# ----------- Time-Frequency Analysis ----------- #
def time_frequency(data, cmwX, nKern):
    nData = data.shape[1]
    nConv = nData + nKern - 1
    halfwav = (nKern - 1) // 2
    tf = np.zeros((data.shape[0], cmwX.shape[0], data.shape[1]))
    for ch in range(data.shape[0]):
        eegX = fft(data[ch, :], nConv)
        as_ = ifft(cmwX * eegX[None, :], axis=1)
        tf[ch, :, :] = np.abs(as_[:, halfwav:-halfwav]) ** 2
    return tf

# ----------- Load Data ----------- #
mnist_path = '/content/drive/MyDrive/MU.txt'
mnist_df = pd.read_csv(mnist_path, sep='\t', header=None, nrows=5000)
mnist_df.columns = ['id', 'event_id', 'device', 'channel', 'code', 'size', 'data']
resting_df = pd.read_csv(mnist_path, sep='\t', names=mnist_df.columns, header=None, skiprows=130000, nrows=5000)
df = pd.concat([mnist_df, resting_df])
df['data'] = df['data'].apply(lambda x: [float(i) for i in x.split(',')])
df['data'] = df['data'].apply(lambda x: bandpass_filter(np.array(x)))

# ----------- Resample ----------- #
median_length = 459
df['resampled_data'] = df['data'].apply(lambda x: resample_array(x, median_length))
data_array = np.array(df['resampled_data'].tolist())
codes = df['code'].tolist()
data_array = data_array.reshape(-1, 4, median_length)
data_array = np.mean(data_array, axis=1)
codes = codes[::4]

# ----------- Time-Frequency Plot Creation ----------- #
times = np.linspace(0, 2, median_length)
cmwX, nKern, frex = get_cmwX(data_array.shape[1])
tf = time_frequency(data_array, cmwX, nKern)

# ----------- Modified get_path function ----------- #
def get_path(i, path, code, total_samples):
    train_ratio = 0.7
    train_cutoff = int(total_samples * train_ratio)
    if i < train_cutoff:
        return f"{path}training/{code}"
    else:
        return f"{path}test/{code}"


```

```python
# ----------- Save Plots ----------- #
path = '/content/drive/MyDrive/multiclass/images/'
total_samples = tf.shape[0]

for i in range(total_samples):
    label = codes[i]  # ✅ Corrected: preserve actual label (-1 to 9)
    folder_path = get_path(i, path, label, total_samples)
    os.makedirs(folder_path, exist_ok=True)
    fig, ax = plt.subplots()
    ax.contourf(times, frex, tf[i, :, :], 40, cmap='jet')
    ax.axis('off')
    fig.savefig(f"{folder_path}/{i}.png", bbox_inches='tight', pad_inches=0)
    plt.close(fig)
```

```python
import os

def count_images_per_class(base_path):
    label_counts = {}

    for label_folder in sorted(os.listdir(base_path)):
        folder_path = os.path.join(base_path, label_folder)
        if os.path.isdir(folder_path):
            num_images = len([f for f in os.listdir(folder_path) if f.endswith('.png')])
            label_counts[label_folder] = num_images

    return label_counts

# Replace with your training and test folder paths:
train_path = '/content/drive/MyDrive/multiclass/images/training'
test_path = '/content/drive/MyDrive/multiclass/images/test'

train_counts = count_images_per_class(train_path)
test_counts = count_images_per_class(test_path)

print("Training image counts per label:")
for label, count in train_counts.items():
    print(f"Label {label}: {count} images")

print("\nTest image counts per label:")
for label, count in test_counts.items():
    print(f"Label {label}: {count} images")
```

```python
import os
import shutil
import random

def balance_test_set(train_path, test_path, target_count=100):
    labels = [str(i) for i in range(10)]  # labels 0 to 9

    for label in labels:
        print(f"\n🔍 Processing label: {label}")

        train_label_path = os.path.join(train_path, label)
        test_label_path = os.path.join(test_path, label)

        if not os.path.exists(train_label_path):
            print(f"❌ Skipped label {label} - training folder does not exist.")
            continue

        os.makedirs(test_label_path, exist_ok=True)

        train_images = [f for f in os.listdir(train_label_path) if f.endswith('.png')]
        test_images = [f for f in os.listdir(test_label_path) if f.endswith('.png')]
        current_test_count = len(test_images)
        needed = target_count - current_test_count

        if needed > 0 and len(train_images) > 0:
            move_count = min(needed, len(train_images))
            to_move = random.sample(train_images, move_count)
            for img in to_move:
                shutil.move(os.path.join(train_label_path, img), os.path.join(test_label_path, img))
            print(f"✅ Moved {move_count} images from training/{label} to test/{label}")
        else:
            print(f"⚠️  Label {label} already has {current_test_count} or not enough images to move.")

# ✅ Call the function
balance_test_set(
    train_path='/content/drive/MyDrive/multiclass/images/training',
    test_path='/content/drive/MyDrive/multiclass/images/test',
    target_count=100  # or any number you'd like per class
)
```

```python
import os
import shutil
import random
from math import ceil

def oversample_training_set(train_path, target_count=None):
    """
    Oversamples classes 1 to 9 in the training set by duplicating images until
    each class has approximately target_count images.

    If target_count is None, set it to max class count in training set.
    """
    labels = [str(i) for i in range(1, 10)]  # Classes 1 to 9

    # Count current images per label
    counts = {}
    for label in labels:
        label_path = os.path.join(train_path, label)
        if not os.path.exists(label_path):
            print(f"Warning: Folder {label_path} does not exist.")
            continue
        counts[label] = len([f for f in os.listdir(label_path) if f.endswith('.png')])

    # Determine target_count if not specified (max count among classes 1-9)
    if target_count is None:
        target_count = max(counts.values())

    print(f"Target oversample count per class: {target_count}")

    for label in labels:
        label_path = os.path.join(train_path, label)
        current_count = counts.get(label, 0)

        if current_count >= target_count:
            print(f"Label {label} already has {current_count} images (>= {target_count}). Skipping.")
            continue

        # Number of images to add
        needed = target_count - current_count
        print(f"Label {label}: current {current_count}, need {needed} more images.")

        images = [f for f in os.listdir(label_path) if f.endswith('.png')]
        if not images:
            print(f"No images found for label {label}. Skipping.")
            continue

        # Repeat copying random images from existing ones until we reach target_count
        for i in range(needed):
            src_img = random.choice(images)
            src_path = os.path.join(label_path, src_img)
            # New filename to avoid overwriting
            dst_img = f"{src_img.split('.')[0]}_dup_{i}.png"
            dst_path = os.path.join(label_path, dst_img)

            shutil.copy(src_path, dst_path)

        print(f"Oversampled label {label} with {needed} duplicated images.")

# Example usage:
train_folder_path = "/content/drive/MyDrive/multiclass/images/training"
oversample_training_set(train_folder_path)
```

```python
from fastai.vision.all import *

path = "/content/drive/MyDrive/multiclass/images"
dls = ImageDataLoaders.from_folder(path, train='training', valid_pct=0.2, item_tfms=Resize(224))
learn = vision_learner(dls, resnet50, metrics=accuracy)

# Fine-tune for a reasonable number of epochs, e.g., 20-30
learn.fine_tune(20, base_lr=3e-4)

# Evaluate on test set
test_path = "/content/drive/MyDrive/multiclass/images/test"
test_files = get_image_files(test_path)
test_dl = learn.dls.test_dl(test_files, with_labels=True)
loss, acc = learn.validate(dl=test_dl)
print(f"Test Accuracy: {acc*100:.2f}%")
```
